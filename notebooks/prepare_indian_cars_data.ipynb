```python
import random
import shutil
from pathlib import Path
```

```python
Base_dir = Path.cwd().parent

Indian_raw = Base_dir/"dataset/Indian_car_data/raw_images"

Train_dir = Base_dir/"notebooks/dataset/processed/train"
```

```python
for brand_dir in Indian_raw.iterdir():
    if not brand_dir.is_dir:
        continue

    for model_dir in brand_dir.iterdir():
        if not model_dir.is_dir:
            continue

        class_name = f"{brand_dir.name} _ {model_dir.name}"

        images = list(model_dir.glob("*jpg")) + \
                 list(model_dir.glob("*png")) + \
                 list(model_dir.glob("*jpeg"))
        
        if len(images) < 5:
            continue

        targer_dir = Train_dir/class_name
        targer_dir.mkdir(parents=True, exist_ok=True)

        for img in images:
            shutil.copy(img, targer_dir/img.name)
```
